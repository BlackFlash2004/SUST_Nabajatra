# অলীকবচন — `qwen_gd_Highest` Kaggle Submit

Kaggle version of the **high-accuracy path** from `qwen_gd_Highest.ipynb`.

Does **not** re-run Qwen / 80k / Bagdhara. It merges the specialist outputs those cells already produced, with the **FINAL MERGE fix** (trust full `subset_label`, not only hits).

| Subset | Source | Gold acc (approx) |
|--------|--------|-------------------|
| **A** | `answer_in_context` + `nli_contra ≤ 0.5` | ~85% |
| **B** | `review_subset_B.csv` → `subset_label` | ~90% |
| **C** | `review_subset_C.csv` → `subset_label` | ~100% |
| **D** | review label, overridden by arithmetic solvers when parseable | ~77–90% |
| **E** | `review_subset_E.csv` → `subset_label` | ~100% |

**Overall vs gold ≈ 89%** (vs ~65% for broken `submission_final`).

```python
# Cell 1 — imports
import os, re, json
from pathlib import Path
from typing import Optional

import numpy as np
import pandas as pd

print("pandas", pd.__version__)
```

```python
# Cell 2 — CONFIG (auto-find under /kaggle/input)
ON_KAGGLE = Path("/kaggle").exists()
SEARCH_ROOTS = [
    Path("/kaggle/input"),
    Path("/kaggle/working"),
    Path("."),
    Path("Subset Files"),
]

def _find_file(*name_parts: str) -> Optional[str]:
    parts = [p.lower() for p in name_parts]
    hits = []
    for base in SEARCH_ROOTS:
        if not base.exists():
            continue
        for h in base.rglob("*"):
            if h.is_file() and all(p in h.name.lower() for p in parts):
                hits.append(h)
    if not hits:
        return None
    hits.sort(key=lambda p: (0 if "review" in p.name.lower() else 1, len(str(p))))
    return str(hits[0])

def _require(*parts, fallback: str) -> str:
    return _find_file(*parts) or fallback

_out_dir = Path("/kaggle/working") if ON_KAGGLE else Path(".")
_out_dir.mkdir(parents=True, exist_ok=True)

# --- Which test set to score -------------------------------------------------
# Set this env var (or just edit the literal) to point at ANY test CSV with the
# same schema (id, context, prompt_bn, response_bn) — the original ~2516-row
# competition test set, or a broader evaluation set (e.g. a 5k-row set). The
# rest of the pipeline (segmentation, D solvers, A/C live grounding) does not
# hard-code row counts or ids anywhere, so it scales to any size automatically.
# Only B/E rows OUTSIDE the original review files fall back to a weaker
# heuristic — see the warning printed after Cell 6/7.
TEST_PATH_OVERRIDE = os.environ.get("HALLU_TEST_PATH", "").strip()

COMP_DIR = "/kaggle/input/competitions/bengali-hallucination"


def _resolve_test_path() -> str:
    if TEST_PATH_OVERRIDE and os.path.exists(TEST_PATH_OVERRIDE):
        return TEST_PATH_OVERRIDE
    # Competition dataset directory takes priority — check it directly first
    # (covers both the original test set and a broader/5k-row replacement dropped
    # into the same mounted directory).
    if Path(COMP_DIR).exists():
        for name in ("test set.csv", "test_set.csv"):
            cand = os.path.join(COMP_DIR, name)
            if os.path.exists(cand):
                return cand
        for parts in (("5k", ".csv"), ("broader", ".csv"), ("test", ".csv")):
            hits = [
                h for h in Path(COMP_DIR).rglob("*")
                if h.is_file() and all(p in h.name.lower() for p in parts)
            ]
            if hits:
                return str(sorted(hits, key=lambda p: len(str(p)))[0])
    # Fall back to the standard local/Colab search (unchanged behavior when the
    # competition directory isn't mounted, e.g. running outside Kaggle).
    for exact in ("test set.csv", "test_set.csv"):
        if os.path.exists(exact):
            return exact
    std_hit = _find_file("test set", ".csv") or _find_file("test_set", ".csv")
    if std_hit:
        return std_hit
    for parts in (("5k", ".csv"), ("broader", ".csv"), ("test", "5000", ".csv")):
        hit = _find_file(*parts)
        if hit:
            return hit
    return "test set.csv"

CONFIG = {
    "test_path": _resolve_test_path(),
    "sample_path": (
        _find_file("dataset samples.json")
        or _find_file("dataset_samples.json")
        or "dataset samples.json"
    ),
    "gold_path": _require(
        "submission_Subset-A_B_C_D_E_corrected", ".csv",
        fallback="submission_Subset-A_B_C_D_E_corrected.csv",
    ),
    "subset_a_features": _require(
        "subset_A_standard_context", ".csv",
        fallback="Subset Files/subset_A_standard_context_test.csv",
    ),
    "review_b": _require("review_subset_B", ".csv", fallback="Subset Files/review_subset_B.csv"),
    "review_c": _require("review_subset_C", ".csv", fallback="Subset Files/review_subset_C.csv"),
    "review_d": _require("review_subset_D", ".csv", fallback="Subset Files/review_subset_D.csv"),
    "review_e": _require("review_subset_E", ".csv", fallback="Subset Files/review_subset_E.csv"),
    "output": str(_out_dir / "submission.csv"),
    "audit_output": str(_out_dir / "submission_audit.csv"),
    "nli_contra_threshold": 0.5,
}

print("ON_KAGGLE:", ON_KAGGLE)
if TEST_PATH_OVERRIDE:
    print(f"TEST_PATH_OVERRIDE set: {TEST_PATH_OVERRIDE}")
for k, v in CONFIG.items():
    if isinstance(v, str) and (v.endswith(".csv") or v.endswith(".json")):
        print(f"  {k}: {v}  [{'OK' if os.path.exists(v) else 'MISSING'}]")
    else:
        print(f"  {k}: {v}")

missing_critical = [
    k for k in ["test_path", "review_b", "review_c", "review_d", "review_e", "subset_a_features"]
    if not os.path.exists(CONFIG[k])
]
if missing_critical:
    print("\n⚠ Missing:", missing_critical)
    print("  Upload Subset Files (review CSVs + A features) as a Kaggle dataset and re-run.")
    if "test_path" in missing_critical:
        print("  To score a NEW/broader test set: set env var HALLU_TEST_PATH to its path,")
        print("  or mount it under /kaggle/input with '5k' or 'broader' in the filename.")
```

```python
# Cell 3 — Load test (+ optional gold / sample)
assert os.path.exists(CONFIG["test_path"]), (
    f"Missing test set: {CONFIG['test_path']}\n"
    "Add competition data, then: !ls -R /kaggle/input | head -80"
)

test = pd.read_csv(CONFIG["test_path"])
test["id"] = test["id"].astype(int)
print(f"Test: {len(test)} rows")

HAS_GOLD = os.path.exists(CONFIG["gold_path"])
gold_map = {}
if HAS_GOLD:
    gold_df = pd.read_csv(CONFIG["gold_path"])
    gold_map = dict(zip(gold_df["id"].astype(int), gold_df["label"].astype(int)))
    print(f"Gold: {len(gold_map)} (offline only)")

if os.path.exists(CONFIG["sample_path"]):
    with open(CONFIG["sample_path"], encoding="utf-8") as f:
        sample_df = pd.DataFrame(json.load(f))
    print(f"Sample: {len(sample_df)}")
else:
    sample_df = pd.DataFrame()
```

```python
# Cell 4 — Segmentation A/B/C/D/E
_BN2AR = str.maketrans("০১২৩৪৫৬৭৮৯", "0123456789")
_CTX_NULL = {"[null]", "null", "none", "nan", ""}

IDIOM_PAT = re.compile(
    r"(?:ভাবার্থ|শাব্দিক\s*অর্থ|শব্দার্থ|বাগধারা|প্রবাদের\s*অর্থ|"
    r"এর\s*অর্থ\s*কী|এর\s*অর্থ\s*কি|অর্থ\s*কী\?|অর্থ\s*কি\?)", re.I)
MATH_PAT = re.compile(
    r"(?:[0-9০-৯]|[\+\-\*/×÷=<>≤≥]|সমাধান\s*সেট|অসমতা|সমীকরণ|"
    r"ক্রয়মূল্য|বিক্রয়মূল্য|সরল\s*সুদ|অনুপাতে\s*বণ্টন|গতিবেগ|"
    r"ত্রিভুজ|ক্ষেত্রফল|একাই|একা[^\n]{0,40}দিনে|শতকরা\s*কত\s*লাভ|%|বার\s*হলে)", re.I)

def is_null_context(ctx) -> bool:
    return str(ctx).strip().lower() in _CTX_NULL

def norm_text(s: str) -> str:
    return re.sub(r"\s+", "", str(s).translate(_BN2AR).lower())

def answer_in_context(a, c) -> bool:
    a, c = str(a).strip(), str(c)
    if not a or is_null_context(c):
        return False
    if a in c:
        return True
    na, nc = norm_text(a), norm_text(c)
    return bool(na) and na in nc

def token_cov(a, c) -> float:
    if is_null_context(c):
        return 0.0
    ta, tc = set(str(a).lower().split()), set(str(c).lower().split())
    return len(ta & tc) / len(ta) if ta else 0.0

def assign_subset_row(row) -> str:
    ctx = str(row.get("context", ""))
    p = str(row.get("prompt_bn", ""))
    a = str(row.get("response_bn", ""))
    if IDIOM_PAT.search(p):
        return "E"
    if is_null_context(ctx):
        return "D" if MATH_PAT.search(p) else "B"
    if not (answer_in_context(a, ctx) or token_cov(a, ctx) >= 0.6):
        return "C"
    return "A"

subset_map = {}
for letter, parts in {
    "A": ("subset_A_standard_context", ".csv"),
    "B": ("subset_B_null_context", ".csv"),
    "C": ("review_subset_C", ".csv"),
    "D": ("subset_D_math", ".csv"),
    "E": ("review_subset_E", ".csv"),
}.items():
    path = _find_file(*parts)
    if not path and letter == "E":
        path = _find_file("subset_E_idiom", ".csv")
    if not path:
        print(f"  {letter}: no partition file — regex later")
        continue
    sub = pd.read_csv(path)
    for rid in sub["id"].astype(int):
        subset_map[int(rid)] = letter
    print(f"  {letter}: {len(sub)} ← {path}")

# reviews also define membership
for letter, key in [("B", "review_b"), ("C", "review_c"), ("D", "review_d"), ("E", "review_e")]:
    path = CONFIG[key]
    if os.path.exists(path):
        for rid in pd.read_csv(path)["id"].astype(int):
            subset_map.setdefault(int(rid), letter)

test["subset"] = [
    subset_map.get(int(i), assign_subset_row(test.loc[test["id"] == i].iloc[0]))
    for i in test["id"]
]
print(test["subset"].value_counts())
```

```python
# Cell 5 — D arithmetic solvers (inline)
from __future__ import annotations

import math
import re
from fractions import Fraction
from typing import Any

_BN_DIGITS = "০১২৩৪৫৬৭৮৯"
_NUM = r"(\d+(?:[,\.]\d+)?|[০-৯]+(?:[,\.][০-৯]+)?)"
_NUM_G = re.compile(_NUM)

WEEKDAYS = [
    "রবিবার",
    "সোমবার",
    "মঙ্গলবার",
    "বুধবার",
    "বৃহস্পতিবার",
    "শুক্রবার",
    "শনিবার",
]
WEEKDAY_INDEX = {w: i for i, w in enumerate(WEEKDAYS)}


def bn_to_num(s: str) -> float:
    out = []
    for c in str(s).replace(" ", ""):
        if c in _BN_DIGITS:
            out.append(str(_BN_DIGITS.index(c)))
        elif c in ",.":
            out.append(c)
        elif c.isdigit():
            out.append(c)
    text = "".join(out).replace(",", "")
    if not text:
        return float("nan")
    return float(text)


def bn_to_int(s: str) -> int:
    return int(round(bn_to_num(s)))


def extract_numbers(text: str) -> list[float]:
    return [bn_to_num(m.group(1)) for m in _NUM_G.finditer(str(text))]


def response_number(resp: str) -> float:
    nums = extract_numbers(resp)
    return nums[0] if nums else float("nan")


def numbers_close(a: float, b: float, tol: float = 0.51) -> bool:
    if math.isnan(a) or math.isnan(b):
        return False
    if abs(a - round(a)) < 1e-9 and abs(b - round(b)) < 1e-9:
        return int(round(a)) == int(round(b))
    return abs(a - b) <= tol


def safe_round(val: float) -> float | None:
    if val is None or math.isnan(val) or math.isinf(val):
        return None
    return round(val)


def weekday_in_text(text: str) -> str | None:
    for w in WEEKDAYS:
        if w in str(text):
            return w
    return None


def weekday_add(start: str, days: int) -> str:
    return WEEKDAYS[(WEEKDAY_INDEX[start] + int(days)) % 7]


# --- work rate ---
WORK2_PAT = re.compile(
    r"একা[^\n]{0,100}?" + _NUM + r"[^\n]{0,40}?দিনে[^\n]{0,140}?"
    + r"একা[^\n]{0,100}?" + _NUM + r"[^\n]{0,40}?দিনে",
    re.I,
)
WORK3_PAT = re.compile(
    r"যথাক্রমে\s*" + _NUM + r"\s*,\s*" + _NUM + r"\s*(?:,\s*|ও\s*)" + _NUM + r"\s*দিনে",
    re.I,
)


def solve_work_rate(prompt: str) -> float | None:
    m2 = WORK2_PAT.search(prompt)
    if m2:
        a, b = bn_to_int(m2.group(1)), bn_to_int(m2.group(2))
        if a > 0 and b > 0:
            return safe_round(1 / (1 / a + 1 / b))
    m3 = WORK3_PAT.search(prompt)
    if m3:
        a, b, c = bn_to_int(m3.group(1)), bn_to_int(m3.group(2)), bn_to_int(m3.group(3))
        if min(a, b, c) > 0:
            return safe_round(1 / (1 / a + 1 / b + 1 / c))
    return None


# --- ratio sum (age / count) ---
RATIO_SUM_PAT = re.compile(
    r"অনুপাত\s*" + _NUM + r"\s*:\s*" + _NUM + r"[^\n]{0,120}?"
    + r"(?:সমষ্টি|সংখ্যা|মোট)[^\n]{0,30}?" + _NUM,
    re.I,
)


def solve_ratio_sum(prompt: str) -> float | None:
    m = RATIO_SUM_PAT.search(prompt)
    if not m:
        return None
    a, b, total = bn_to_num(m.group(1)), bn_to_num(m.group(2)), bn_to_num(m.group(3))
    part = _ratio_part_for_question(prompt, a, b)
    if part is None:
        return None
    return safe_round(total * part / (a + b))


def _ratio_part_for_question(prompt: str, a: float, b: float) -> float | None:
    ask = prompt.split("?", 1)[0]
    second_kw = ("ছাগল", "কাতলা", "পানি", "মেয়ে", "ছোট")
    first_kw = ("গরু", "রুই", "চিনি", "মা", "বড়")
    if any(k in ask for k in second_kw):
        return b
    if any(k in ask for k in first_kw):
        return a
    head = prompt.split("অনুপাত", 1)[0]
    entities = re.split(r"\s+ও\s+|\s+এবং\s+", head)
    if len(entities) >= 2:
        e1, e2 = entities[-2], entities[-1]
        for w in re.findall(r"[^\s,]+", e2):
            if len(w) > 2 and w in ask:
                return b
        for w in re.findall(r"[^\s,]+", e1):
            if len(w) > 2 and w in ask:
                return a
    return b


# --- ratio partition among partners ---
PARTITION_PAT = re.compile(
    _NUM + r"\s*টাকা[^\n]{0,100}?" + _NUM + r"\s*:\s*" + _NUM + r"\s*:\s*" + _NUM + r"\s*অনুপাত",
    re.I,
)


def solve_partition(prompt: str) -> float | None:
    m = PARTITION_PAT.search(prompt)
    if not m:
        return None
    total = bn_to_num(m.group(1))
    r1, r2, r3 = bn_to_num(m.group(2)), bn_to_num(m.group(3)), bn_to_num(m.group(4))
    ratios = [r1, r2, r3]
    idx = 1
    if re.search(r"প্রথম\s*অংশীদার", prompt):
        idx = 0
    elif re.search(r"তৃতীয়\s*অংশীদার", prompt):
        idx = 2
    s = sum(ratios)
    return safe_round(total * ratios[idx] / s)


# --- profit / loss ---
PROFIT_PAT = re.compile(
    _NUM + r"[^\n]{0,40}?টাক[^\n]{0,80}?" + _NUM + r"\s*%\s*(লাভ|ক্ষতি)",
    re.I,
)
TWO_STEP_PAT = re.compile(
    r"প্রথম[^\n]{0,40}?" + _NUM + r"\s*%\s*(বৃদ্ধি|বেড়ে)[^\n]{0,80}?"
    + r"(?:দ্বিতীয়|পরে)[^\n]{0,40}?" + _NUM + r"\s*%\s*(ছাড়|কমে)[^\n]{0,80}?"
    + r"(?:শুরুর|প্রাথমিক)\s*দাম\s*" + _NUM,
    re.I,
)


def solve_profit_loss(prompt: str) -> float | None:
    m2 = TWO_STEP_PAT.search(prompt)
    if m2:
        p1, p2, base = bn_to_num(m2.group(1)), bn_to_num(m2.group(2)), bn_to_num(m2.group(3))
        val = base * (1 + p1 / 100) * (1 - p2 / 100)
        return safe_round(val)
    m = PROFIT_PAT.search(prompt)
    if not m:
        return None
    price, pct, kind = bn_to_num(m.group(1)), bn_to_num(m.group(2)), m.group(3)
    if kind == "লাভ":
        return safe_round(price * (1 + pct / 100))
    return safe_round(price * (1 - pct / 100))


# --- simple interest ---
SI_PAT = re.compile(
    _NUM + r"\s*টাকা[^\n]{0,60}?বার্ষিক\s*" + _NUM + r"\s*%\s*সরল\s*সুদ[^\n]{0,40}?"
    + r"(\d+|[০-৯]+)\s*বছর",
    re.I,
)
SI_PAT2 = re.compile(
    _NUM + r"\s*টাকা\s*(\d+|[০-৯]+)\s*বছর[^\n]{0,60}?বার্ষিক\s*" + _NUM + r"\s*%\s*সরল\s*সুদ",
    re.I,
)


def solve_simple_interest(prompt: str) -> float | None:
    m = SI_PAT.search(prompt) or SI_PAT2.search(prompt)
    if not m:
        return None
    p, r, t = bn_to_num(m.group(1)), bn_to_num(m.group(2)), bn_to_int(m.group(3))
    return safe_round(p * r * t / 100)


# --- speed / distance ---
GAP_PAT = re.compile(
    r"গতিবেগ\s*ঘণ্টায়\s*" + _NUM + r"\s*কিমি[^\n]{0,80}?ঘণ্টায়\s*" + _NUM
    + r"\s*কিমি[^\n]{0,40}?(\d+|[০-৯]+)\s*ঘণ্টা",
    re.I,
)
MEET_PAT = re.compile(
    r"দূরত্ব\s*" + _NUM + r"\s*কিলোমিটার[^\n]{0,80}?ঘণ্টায়\s*" + _NUM
    + r"\s*কিমি[^\n]{0,80}?ঘণ্টায়\s*" + _NUM + r"\s*কিমি",
    re.I,
)


def solve_speed(prompt: str) -> float | None:
    m = GAP_PAT.search(prompt)
    if m:
        v1, v2, t = bn_to_num(m.group(1)), bn_to_num(m.group(2)), bn_to_int(m.group(3))
        return safe_round(abs(v1 - v2) * t)
    m = MEET_PAT.search(prompt)
    if m:
        d, v1, v2 = bn_to_num(m.group(1)), bn_to_num(m.group(2)), bn_to_num(m.group(3))
        if v1 + v2 > 0:
            val = d / (v1 + v2)
            return round(val, 2) if not math.isnan(val) else None
    return None


# --- LCM ---
LCM_PAT = re.compile(
    r"যথাক্রমে\s*প্রতি\s*" + _NUM + r"\s*,\s*" + _NUM + r"\s*(?:,\s*|ও\s*)" + _NUM + r"\s*মিনিট",
    re.I,
)


def _lcm3(a: int, b: int, c: int) -> int:
    return math.lcm(math.lcm(a, b), c)


def solve_lcm(prompt: str) -> float | None:
    m = LCM_PAT.search(prompt)
    if not m:
        return None
    a, b, c = bn_to_int(m.group(1)), bn_to_int(m.group(2)), bn_to_int(m.group(3))
    if min(a, b, c) <= 0:
        return None
    return float(_lcm3(a, b, c))


# --- combinations ---
COMB_PAT = re.compile(
    r"(\d+|[০-৯]+)\s*জন[^\n]{0,80}?(\d+|[০-৯]+)\s*জন",
    re.I,
)


def solve_combination(prompt: str) -> float | None:
    m = COMB_PAT.search(prompt)
    if not m or "প্যানেল" not in prompt and "উপকমিটি" not in prompt:
        return None
    n, k = bn_to_int(m.group(1)), bn_to_int(m.group(2))
    if k > n or k < 0:
        return None
    return float(math.comb(n, k))


# --- average (new value) ---
AVG_PAT = re.compile(
    r"(\d+|[০-৯]+)\s*টি\s*রাশির\s*গড়মান\s*(\d+|[০-৯]+)[^\n]{0,80}?গড়\s*দাঁড়ায়\s*(\d+|[০-৯]+)",
    re.I,
)
CLASS_AVG_PAT = re.compile(
    r"(\d+|[০-৯]+)\s*জন\s*শিক্ষার্থীর\s*গড়\s*নম্বর\s*(\d+|[০-৯]+)[^\n]{0,80}?গড়\s*নম্বর\s*হয়\s*(\d+|[০-৯]+)",
    re.I,
)


def solve_average(prompt: str) -> float | None:
    m = AVG_PAT.search(prompt)
    if m:
        n, old, new = bn_to_int(m.group(1)), bn_to_num(m.group(2)), bn_to_num(m.group(3))
        return safe_round(new * (n + 1) - old * n)
    m = CLASS_AVG_PAT.search(prompt)
    if m:
        n, old, new = bn_to_int(m.group(1)), bn_to_num(m.group(2)), bn_to_num(m.group(3))
        return safe_round(new * (n + 1) - old * n)
    return None


# --- percent reverse ---
PCT_REV_PAT = re.compile(
    r"(\d+|[০-৯]+)\s*এর\s*কত\s*শতাংশ\s*(\d+|[০-৯]+)\s*হবে",
    re.I,
)
FRAC_PAT = re.compile(
    r"(\d+|[০-৯]+)/(\d+|[০-৯]+)\s*অংশ\s*(\d+|[০-৯]+)",
    re.I,
)


def solve_percent(prompt: str) -> float | None:
    m = PCT_REV_PAT.search(prompt)
    if m:
        whole, part = bn_to_num(m.group(1)), bn_to_num(m.group(2))
        if whole:
            return safe_round(100 * part / whole)
    m = FRAC_PAT.search(prompt)
    if m:
        num, den, val = bn_to_int(m.group(1)), bn_to_int(m.group(2)), bn_to_num(m.group(3))
        if den:
            return safe_round(val * den / num)
    return None


# --- unit price ---
UNIT_PAT = re.compile(
    r"প্রতি[^\n]{0,20}?" + _NUM + r"\s*টাক[^\n]{0,40}?" + _NUM + r"\s*টাক[^\n]{0,40}?কত",
    re.I,
)


def solve_unit_price(prompt: str) -> float | None:
    m = UNIT_PAT.search(prompt)
    if not m:
        return None
    unit, budget = bn_to_num(m.group(1)), bn_to_num(m.group(2))
    if unit > 0:
        return safe_round(budget / unit)
    return None


# --- calendar ---
CAL_TODAY_PAT = re.compile(
    r"আজ\s*যদি\s*(\S+বার)\s*হয়[^\n]{0,40}?(\d+|[০-৯]+)\s*দিন\s*পরে",
    re.I,
)
CAL_DAY_PAT = re.compile(
    r"(\S+বার)\s*বার\s*হলে[^\n]{0,40}?(\d+|[০-৯]+)\s*দিন\s*পর",
    re.I,
)


def solve_calendar(prompt: str) -> str | None:
    m = CAL_TODAY_PAT.search(prompt) or CAL_DAY_PAT.search(prompt)
    if not m:
        return None
    day_raw, nd = m.group(1), bn_to_int(m.group(2))
    day = day_raw if day_raw in WEEKDAY_INDEX else weekday_in_text(day_raw)
    if not day:
        return None
    return weekday_add(day, nd)



# --- reverse percentage (price up -> reduce quantity to keep spend flat) ---
REV_PCT_PAT = re.compile(
    r"(\d+|[০-৯]+)\s*%\s*(বৃদ্ধি|বেড়ে)[^\n]{0,120}?(ব্যয়|খরচ|খরচা)[^\n]{0,40}?(অপরিবর্তিত|একই|সমান)",
    re.I,
)


def solve_reverse_percent(prompt: str) -> float | None:
    m = REV_PCT_PAT.search(prompt)
    if not m:
        return None
    pct = bn_to_num(m.group(1))
    if pct <= -100:
        return None
    # reduction% = pct / (100+pct) * 100
    return safe_round(pct / (100 + pct) * 100)


# --- compound interest: amount = P(1+r/100)^n ---
CI_PAT = re.compile(
    r"(\d+|[০-৯]+)\s*টাকা[^\n]{0,60}?(চক্রবৃদ্ধি)[^\n]{0,60}?(\d+|[০-৯]+)\s*%[^\n]{0,60}?(\d+|[০-৯]+)\s*বছর",
    re.I,
)


def solve_compound_interest(prompt: str) -> float | None:
    m = CI_PAT.search(prompt)
    if not m:
        return None
    p, rate, n = bn_to_num(m.group(1)), bn_to_num(m.group(3)), bn_to_int(m.group(4))
    return safe_round(p * (1 + rate / 100) ** n)


# --- Heron's formula: triangle area from three sides ---
HERON_PAT = re.compile(
    r"বাহু[^\n]{0,15}?(\d+|[০-৯]+)[^\n]{0,15}?(\d+|[০-৯]+)[^\n]{0,15}?(\d+|[০-৯]+)[^\n]{0,60}?ক্ষেত্রফল",
    re.I,
)


def solve_heron(prompt: str) -> float | None:
    m = HERON_PAT.search(prompt)
    if not m:
        return None
    a, b, c = bn_to_num(m.group(1)), bn_to_num(m.group(2)), bn_to_num(m.group(3))
    s = (a + b + c) / 2
    val = s * (s - a) * (s - b) * (s - c)
    if val < 0:
        return None
    return safe_round(math.sqrt(val))


# --- prime count in an inclusive range ---
RANGE_PRIME_PAT = re.compile(
    r"(\d+|[০-৯]+)\s*(?:থেকে|হতে)\s*(\d+|[০-৯]+)\s*(?:এর\s*মধ্যে|পর্যন্ত)[^\n]{0,40}?মৌলিক\s*সংখ্যা",
    re.I,
)


def _is_prime(n: int) -> bool:
    if n < 2:
        return False
    if n < 4:
        return True
    if n % 2 == 0:
        return False
    i = 3
    while i * i <= n:
        if n % i == 0:
            return False
        i += 2
    return True


def solve_prime_count(prompt: str) -> float | None:
    m = RANGE_PRIME_PAT.search(prompt)
    if not m:
        return None
    lo, hi = bn_to_int(m.group(1)), bn_to_int(m.group(2))
    if lo > hi:
        lo, hi = hi, lo
    return float(sum(1 for n in range(lo, hi + 1) if _is_prime(n)))


# --- polygon interior-angle sum, in "right angles" (90-degree units) ---
POLY_ANGLE_PAT = re.compile(
    r"(\d+|[০-৯]+)\s*(?:বাহু|কোণ)[^\n]{0,40}?(বহুভুজ|ভুজ)[^\n]{0,60}?সমকোণ",
    re.I,
)


def solve_polygon_angle_sum_right_angles(prompt: str) -> float | None:
    m = POLY_ANGLE_PAT.search(prompt)
    if not m:
        return None
    n = bn_to_int(m.group(1))
    if n < 3:
        return None
    total_deg = (n - 2) * 180
    return safe_round(total_deg / 90)


# --- rectangle: perimeter + width:length ratio -> area ---
RECT_RATIO_PAT = re.compile(
    r"(প্রস্থ|দৈর্ঘ্য)[^\n]{0,10}?(\d+|[০-৯]+)\s*/\s*(\d+|[০-৯]+)[^\n]{0,80}?পরিসীমা\s*(\d+|[০-৯]+)",
    re.I,
)


def solve_rect_ratio_area(prompt: str) -> float | None:
    m = RECT_RATIO_PAT.search(prompt)
    if not m:
        return None
    num, den = bn_to_num(m.group(2)), bn_to_num(m.group(3))
    perim = bn_to_num(m.group(4))
    if den == 0:
        return None
    # perim = 2(L+W); W = (num/den) * L  ->  L(1 + num/den) = perim/2
    L = (perim / 2) / (1 + num / den)
    W = L * num / den
    return safe_round(L * W)


SOLVERS: list[tuple[str, Any]] = [
    ("work_rate", solve_work_rate),
    ("ratio_sum", solve_ratio_sum),
    ("partition", solve_partition),
    ("profit_loss", solve_profit_loss),
    ("simple_interest", solve_simple_interest),
    ("compound_interest", solve_compound_interest),
    ("reverse_percent", solve_reverse_percent),
    ("speed", solve_speed),
    ("lcm", solve_lcm),
    ("combination", solve_combination),
    ("average", solve_average),
    ("percent", solve_percent),
    ("unit_price", solve_unit_price),
    ("calendar", solve_calendar),
    ("heron", solve_heron),
    ("prime_count", solve_prime_count),
    ("polygon_angle_sum", solve_polygon_angle_sum_right_angles),
    ("rect_ratio_area", solve_rect_ratio_area),
]


def solve_d(prompt: str) -> tuple[Any, str] | None:
    for name, fn in SOLVERS:
        val = fn(prompt)
        if val is not None:
            return val, name
    return None


def response_matches(expected: Any, response: str) -> bool:
    if isinstance(expected, str):
        got = weekday_in_text(response)
        return got == expected
    got = response_number(response)
    return numbers_close(float(expected), got)


def solve_d_and_match(subset: str, prompt: str, response: str) -> int | None:
    if subset != "D":
        return None
    solved = solve_d(prompt)
    if solved is None:
        return None
    expected, _ = solved
    return 1 if response_matches(expected, response) else 0


def solve_d_row(row):
    if str(row.get("subset", "")) != "D":
        return None
    return solve_d_and_match("D", row.get("prompt_bn", ""), row.get("response_bn", ""))

_d = test[test["subset"] == "D"]
fired = int(_d.apply(solve_d_row, axis=1).notna().sum()) if len(_d) else 0
print(f"D solvers parseable: {fired}/{len(_d)}")
```

```python
# Cell 6 — Load specialist review files (from qwen_gd_Highest)
def _load_csv(path):
    if not os.path.exists(path):
        print(f"MISSING: {path}")
        return pd.DataFrame()
    df = pd.read_csv(path)
    if "id" in df.columns:
        df["id"] = df["id"].astype(int)
    print(f"Loaded {path} ({len(df)})")
    return df

a_feats = _load_csv(CONFIG["subset_a_features"])
rev_b = _load_csv(CONFIG["review_b"])
rev_c = _load_csv(CONFIG["review_c"])
rev_d = _load_csv(CONFIG["review_d"])
rev_e = _load_csv(CONFIG["review_e"])

def _label_map(rev: pd.DataFrame) -> dict:
    if rev.empty or "subset_label" not in rev.columns:
        return {}
    out = {}
    for _, r in rev.iterrows():
        try:
            v = int(float(r["subset_label"]))
            if v in (0, 1):
                out[int(r["id"])] = v
        except Exception:
            pass
    return out

lab_b, lab_c, lab_d, lab_e = map(_label_map, (rev_b, rev_c, rev_d, rev_e))
print(f"Review labels: B={len(lab_b)} C={len(lab_c)} D={len(lab_d)} E={len(lab_e)}")
assert len(lab_b) and len(lab_c) and len(lab_d) and len(lab_e), (
    "Need all four review_subset_*.csv files. Upload Subset Files as a Kaggle dataset."
)
```

## Merge logic (fix vs broken `submission_final`)

`qwen_gd_Highest` FINAL MERGE only applied B/E **hits** and left CatBoost for the rest → ~65% gold.

Here we trust **full** `subset_label` for every B/C/D/E row, then apply D solvers when they fire.


```python
# Cell 7 — Optimal merge (+ D solver overrides)
NLI_CONTRA_THR = CONFIG["nli_contra_threshold"]
a_by_id = a_feats.set_index("id").to_dict("index") if not a_feats.empty else {}

final_labels = np.zeros(len(test), dtype=int)
sources = np.array(["fallback"] * len(test), dtype=object)
label_maps = {"B": lab_b, "C": lab_c, "D": lab_d, "E": lab_e}

for pos, (_, row) in enumerate(test.iterrows()):
    rid = int(row["id"])
    subset = str(row["subset"])

    if subset == "A":
        feat = a_by_id.get(rid, {})
        try:
            aic = int(float(feat.get("answer_in_context", 0) or 0))
            contra = float(feat.get("nli_contra", 0) or 0)
        except Exception:
            aic, contra = 0, 0.0
        if aic and contra <= NLI_CONTRA_THR:
            final_labels[pos], sources[pos] = 1, "A_grounded"
        elif aic:
            final_labels[pos], sources[pos] = 0, "A_contra_reject"
        elif answer_in_context(row["response_bn"], row["context"]):
            final_labels[pos], sources[pos] = 1, "A_live_grounded"
        else:
            final_labels[pos], sources[pos] = 0, "A_not_grounded"
        continue

    # B/C/D/E — full specialist label
    lab_map = label_maps.get(subset, {})
    if rid in lab_map:
        final_labels[pos] = int(lab_map[rid])
        sources[pos] = f"{subset}_review"
    elif not is_null_context(row["context"]):
        final_labels[pos] = 1 if answer_in_context(row["response_bn"], row["context"]) else 0
        sources[pos] = "fallback_ctx"
    else:
        final_labels[pos] = 0
        sources[pos] = "fallback_null"

    # D solver override when parseable (high precision)
    if subset == "D":
        sol = solve_d_row(row)
        if sol is not None:
            final_labels[pos] = int(sol)
            sources[pos] = "D_solver"
    # Safety net: an UNREVIEWED, null-context "B" row that the deterministic D
    # solvers can actually parse is very likely a math word-problem that the
    # regex segmentation missed (common when a broader test set phrases things
    # slightly differently) — verify it exactly instead of leaving the blind
    # fallback_null default.
    elif subset == "B" and sources[pos] == "fallback_null":
        solved = solve_d(str(row.get("prompt_bn", "")))
        if solved is not None:
            expected, _name = solved
            final_labels[pos] = 1 if response_matches(expected, str(row.get("response_bn", ""))) else 0
            sources[pos] = "B_as_D_solver"

test["pred_label"] = final_labels
test["label_source"] = sources

print("Predictions by subset:")
print(pd.crosstab(test["subset"], test["pred_label"]))
print("\nSources:")
print(test["label_source"].value_counts())
print("Overall:", test["pred_label"].value_counts().sort_index().to_dict())

# --- Coverage warning for a broader/new test set ------------------------------
# A and D always use a live/deterministic fallback (context-containment / the D
# solvers), so they generalize to ANY row regardless of review-file coverage.
# B (closed-book GK) and E (idioms) have NO live fallback here — an id missing
# from review_subset_B.csv / review_subset_E.csv silently gets a blind default
# (0 for B, whatever the E cell defaults to) rather than a real verification.
_unreviewed_mask = test["label_source"].isin(["fallback_null", "fallback_ctx", "fallback"])
_unreviewed_be = test[_unreviewed_mask & test["subset"].isin(["B", "E"])]
n_unreviewed = len(_unreviewed_be)
if n_unreviewed:
    print(
        f"\n⚠ {n_unreviewed} subset-B/E row(s) have NO specialist review coverage "
        "(likely new ids not in the original run) and got a blind default label "
        "instead of a real verification. This is expected when scoring a broader "
        "test set with the SAME review_subset_B/E.csv from the original run."
    )
    _needs_path = str(_out_dir / "needs_specialist_regen.csv")
    _unreviewed_be[["id", "subset", "context", "prompt_bn", "response_bn"]].to_csv(
        _needs_path, index=False
    )
    print(f"  Saved the list of these ids to: {_needs_path}")
    print(
        "  To properly cover them: re-run the Subset B (80k QA / TituML lookup) and "
        "Subset E (Bagdhara idiom lookup) cells from the full qwen_gd_Highest notebook "
        "on just these rows, then merge the results back into review_subset_B.csv / "
        "review_subset_E.csv before re-running this notebook."
    )
else:
    print("\nAll subset-B/E rows have specialist review coverage — no coverage gap.")
```

```python
# Cell 8 — Offline gold validation (optional)
if HAS_GOLD:
    from sklearn.metrics import accuracy_score, f1_score, classification_report, confusion_matrix

    y_true = test["id"].map(gold_map)
    mask = y_true.notna()
    yt = y_true[mask].astype(int).values
    yp = test.loc[mask, "pred_label"].astype(int).values

    print(f"accuracy={accuracy_score(yt, yp):.4f}")
    print(f"macro-F1={f1_score(yt, yp, average='macro'):.4f}")
    print(confusion_matrix(yt, yp))
    print(classification_report(yt, yp, target_names=["hall(0)", "faith(1)"]))
    print("\nPer-subset:")
    for s in sorted(test["subset"].unique()):
        m = mask & (test["subset"] == s)
        if m.sum() == 0:
            continue
        print(f"  {s}: n={int(m.sum())} acc={accuracy_score(test.loc[m,'id'].map(gold_map), test.loc[m,'pred_label']):.4f}")

    old_path = _find_file("submission_final", ".csv") or "submission_final.csv"
    if os.path.exists(old_path):
        old = pd.read_csv(old_path)
        old_map = dict(zip(old["id"].astype(int), old["label"].astype(int)))
        y_old = test.loc[mask, "id"].map(old_map).astype(int).values
        print(f"\nsubmission_final acc={accuracy_score(yt, y_old):.4f}")
        print(f"gain={accuracy_score(yt, yp) - accuracy_score(yt, y_old):+.4f}")
else:
    print("No gold file — skip validation")
```
